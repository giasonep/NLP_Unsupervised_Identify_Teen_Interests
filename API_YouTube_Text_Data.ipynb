```python
import pandas as pd
import json
import time 

from youtube_transcript_api import YouTubeTranscriptApi
# # displays a YouTube video in Jupyter Notebook
# from IPython.display import YouTubeVideo
```

```python
url_file =  open('url_list.txt', 'r') 
url_list = []
for line in url_file:
    stripped_line = line.strip()
    url_list.append(stripped_line)
```

```python
video_id_list = []
for i in url_list:
    video_id_list.append(i.split('=', maxsplit=1)[1])
```

```python
video_text_data = []
```

```python
# assign count
count = 0

# start time to calculate the time it takes for the script to run
start_time = time.time()

# for look to run though each url and pull its transcript from the API
for vidId in video_id_list:
    count += 1
    try:
        transcript = YouTubeTranscriptApi.get_transcript(vidId)
        textBlock = ''
        for i in transcript:
            textBlock += ' ' + i['text']
        video_text_data.append({"vidId": vidId,
                                "videoText": textBlock,})

        print('Count:', count, 'vidId:', vidId)
    except:
        print('Count:', count, "...DID NOT PULL...")

print("\n\nDONE! This script took", round((time.time() - start_time)/60, 2), "minutes to run.")
```

```python
df = pd.DataFrame(video_text_data) 
print(df.shape)
df.tail(10)
```

```python
# export the dataframe to a csv 
df.to_csv('youtube_text_data.csv', index=False)
```
